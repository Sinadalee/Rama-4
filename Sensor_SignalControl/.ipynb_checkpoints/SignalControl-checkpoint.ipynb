```python
import os, sys
sys.path.append(os.path.join('/home/ring/sumo-svn/', 'tools'))
import traci
import traci.constants as tc
# import numpy as np
from sumolib import checkBinary
import datetime
from datetime import timedelta
# import matplotlib
# import matplotlib.pyplot as plt
# import copy
# import csv

import pandas as pd
```

```python
#reset the environment
def reset():
    sumoBinary = checkBinary('sumo-gui')
    traci.start([sumoBinary, "-c", "../Onnut/onnut.sumocfg",
                             "--tripinfo-output", "tripinfo.xml", '--start','true','--quit-on-end','true','--time-to-teleport','-1'])
```

```python
#open simulation
reset()
```

```python
#readdata
filename_sig = 'SignalControl_SampleData.csv'
filename_junction = 'JunctionData.csv'
df_sig = pd.read_csv(filename_sig)
df_junction = pd.read_csv(filename_junction)
```

```python
detector_id = traci.lanearea.getIDList()
detector_count_vehicleid = [ set() for e in detector_id]
columm_detector_count = ['Time']+[e for e in detector_id]
detector_count = pd.DataFrame(columns = columm_detector_count)
detector_count_before = [0 for e in detector_id]

while True:

    print(str(timedelta(seconds=traci.simulation.getTime()))) #get currenttime in simulation
    
    #dataframe which Time = currenttime
    df_sig_time = df_sig[df_sig.Time == str(timedelta(seconds=traci.simulation.getTime()))]
    
    if len(df_sig_time) != 0: #if data available
        for index, row in df_sig_time.iterrows():
            junctionname = row['JunctionName']
            junctionid = df_junction[df_junction['JunctionName'] == junctionname].JunctionID.to_string()[5:] #get junction id from df_junction
            newphase = row['Phase']
            
            #change trafic signal control
            traci.trafficlight.setPhase(junctionid, newphase)
            print('     '+junctionname+":"+ str(newphase))
    
    #output of detector
    

    for id in detector_id:
        id_index = detector_id.index(id)
        detector_count_vehicleid[id_index] = detector_count_vehicleid[id_index].union(traci.lanearea.getLastStepVehicleIDs(id))
        

    if traci.simulation.getTime() % 5 == 0:
        Time_str_minus5 = str(timedelta(seconds=traci.simulation.getTime()-5))
        detector_count_after = [len(e) for e in detector_count_vehicleid]
        
        detector_count_add   = []
        for i in range(len(detector_count_before)):
            detector_count_add.append(detector_count_after[i]-detector_count_before[i])
        detector_count_add = [Time_str_minus5] + detector_count_add
        df = pd.DataFrame([detector_count_add], columns = columm_detector_count)
        detector_count = detector_count.append(df)
        
        detector_count_before = [len(e) for e in detector_count_vehicleid]
 
    # traci step simulation 1 second
    traci.simulationStep()

    if Time_str_minus5 == '17:10:00': break
```

```python
traci.close()
```
