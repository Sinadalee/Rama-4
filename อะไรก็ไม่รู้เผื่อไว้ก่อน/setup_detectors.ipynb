```python
def setUpLaneAreaDetector(POI):

    root = ET.Element('additionals')
    net = sumolib.net.readNet('KASEMAREA/AddLane_edit2.net.xml')

    for key in POI:
        edgeID = net.getEdge(key[:-2])
        numOfLane=edgeID.getLaneNumber()
        edgeLength=edgeID.getLength()
        initial_Pos=0.00
        subelement=ET.Element('e2Detector')
        subelement.set('id',key)
        subelement.set('lane',key)
        subelement.set('pos',str(initial_Pos))
        subelement.set('endPos','0.00')
        subelement.set('friendlyPos','True')
        subelement.set('length',str(edgeLength))
        subelement.set('freq','2.00')
        subelement.set('file',"map_grid_lock_output1.xml")
        subelement.set('cont','0')
        subelement.set('timeThreshold','1.00')
        subelement.set('speedThreshold','1.39')
        subelement.set('jamThreshold','10.00')

        root.append(subelement)

    tree=ET.ElementTree(root)
    filename = "KASEMAREA/detectors_fromNameLane.add.xml"
    tree.write(filename, xml_declaration=True, pretty_print=True)
```

```python
detector_edit_length = {'824116549#0': [0,40,45,-1], '482384850#2': [0,45,50,270,275,300],'824116555#0': [0,135]} #[start, length]
def Kasem_police_setUpLaneAreaDetectorfromLANE(POI):

    root = ET.Element('additionals')
    net = sumolib.net.readNet('Kasemrat-SUMO-UsingBookNetFile/AddLane_edit2.net.xml')

    for key in POI:
        if key[-2] == '_':
            edgeID = net.getEdge(key[:-2])
        else:
            edgeID = net.getEdge(key)
        edgeLength=edgeID.getLength()
        init_pos = []
        length = []
        name = []
        if key[:-2] in detector_edit_length.keys():
            for i in range(len(detector_edit_length[key[:-2]])//2):
                init_pos.append(detector_edit_length[key[:-2]][2*i])
                length.append(detector_edit_length[key[:-2]][2*i+1]-detector_edit_length[key[:-2]][2*i])
                name.append('D'+key+'.'+str(i))
                print(ID.loc[str(key),'name']+'.'+str(i), '=' , 'D'+key+'.'+str(i))
        else :
            init_pos = [0]
            length = [edgeLength]
            name = ['D'+key]
        for i in range(len(name)):
            init_Pos = init_pos[i]
            subelement=ET.Element('e2Detector')
            subelement.set('id',name[i])
            subelement.set('lane',key)
            subelement.set('pos',str(init_Pos))
            subelement.set('friendlyPos','True')
            subelement.set('length',str(length[i]))
            subelement.set('freq','2.00')
            subelement.set('file',"map_grid_lock_output1.xml")
            subelement.set('cont','0')
            subelement.set('timeThreshold','1.00')
            subelement.set('speedThreshold','1.39')
            subelement.set('jamThreshold','10.00')

            root.append(subelement)

    tree=ET.ElementTree(root)
    filename = "Kasemrat-SUMO-UsingBookNetFile/detectors_frompolice_extend.add.xml"
    tree.write(filename, xml_declaration=True, pretty_print=True)
```
